```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Datasets/question_answer5.csv')
df.head()  
```

```python
del df["Unnamed: 0"]
del df["tags"]
df.head(3)
```

```python
import os
os.environ['OPENAI_API_KEY'] = "API Key"
```

```python
df['questions']=df['questions'] + "\n\nAnswers:\n\n"
df['answers']= " "+df['answers'] + " END"
df.head()
```

```python
df.columns =['prompt', 'completion']
```

```python
df = df.to_json()
df
```

```python
prompt=df
```

```python
def generate_reply(df):
  openai.api_key ='API Key'
  response = openai.Completion.create(
    model="text-davinci-003",
    prompt=df,
    temperature=0,
    max_tokens=150,
    top_p=1.0,
    frequency_penalty=0.0,
    presence_penalty=0.0
    )
  return response.choices[0].text
```

```python
concern  = "Good Morning,How are you?\n\nAnswers:\n\n"
```

```python
reply = generate_reply(concern)
print(reply)
```
